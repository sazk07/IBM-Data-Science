```python
import pandas as pd
import matplotlib as mpl
```

```python
df = pd.read_csv("https://cocl.us/datascience_survey_data", index_col=0)
print(df)
```

```python
df.sort_values(by=["Very interested"], inplace=True, ascending=False)
df
```

```python
df2 = 100 * df / 2233
df2 = df2.round(2)
df2
```

```python
ax = df2.plot(
    kind="bar",
    figsize=(20, 8),
    width=0.8,
    color=("#5cb85c", "#5bc0de", "#d9534f"),
    fontsize=14,
)

ax.set_title("Percentage of Respondents' Interest in Data Science Areas", fontsize=16)

ax.legend(fontsize=14)

ax.spines["top"].set_visible(False)
ax.spines["right"].set_visible(False)
ax.spines["left"].set_visible(False)

for perc in ax.patches:
    ax.annotate(
        "{:.2f}".format(perc.get_height()) + "%",
        xy=(perc.get_x() + 0.05, perc.get_height() + 0.80),
    )
```
